```python
# INDIVIDUAL VISUALIZATON 
# Spotify Youtube Dataset  

#This data set was obtained from Kaggle. Here is the link to the [dataset](https://www.kaggle.com/datasets/salvatorerastelli/spotify-and-youtube). 
#The license to this dataset is listed as CC0: Public Domain. This means that anyone can use this dataset without permission as long as it is not for commerical purposes. 
#The dataset is 30.78 MB.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

spotify_youtube = pd.read_csv("Spotify_Youtube.csv")
spotify_youtube
```

```python
#Which songs have the most and least views on Spotify/Youtube?

#Youtube
sorted_views_df = spotify_youtube.sort_values("Views", ascending=False)
most_viewed = sorted_views_df.iloc[0]
least_viewed = sorted_views_df.iloc[-1]

#Spotify 
sorted_streams_df = spotify_youtube.sort_values("Stream", ascending=False)
most_streamed = sorted_streams_df.iloc[0]
least_streamed = sorted_streams_df.iloc[-1]

print('Most viewed song on Youtube: ', most_viewed.Artist, most_viewed['Title'])
print('Least viewed song on Youtube: ', least_viewed.Artist, least_viewed['Title'])
print('Most streamed song on Spotify: ', most_streamed['Artist'], most_streamed['Title'])
print('Least streamed song on Spotify: ', least_streamed['Artist'], least_streamed['Title'])
```

```python
sns.lmplot(x="Stream", y="Energy", data=sorted_streams_df, line_kws={"color": "C1"})
plt.title("Do higher engery songs get more streams?")
```

```python
sns.lmplot(x="Stream", y="Danceability", data=sorted_streams_df, line_kws={"color": "C1"})
plt.title("Do more Danceable songs get more streams?")
```

```python
#Average duration of different artists and their songs 

#Metallica
metallica_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Metallica']
avg_metallica_duration = metallica_df['Duration_ms'].mean()

#Zendaya
Zendaya_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Zendaya']
avg_Zendaya_duration = Zendaya_df['Duration_ms'].mean()

#Of Monsters and Men
omm_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Of Monsters and Men']
avg_omm_duration = omm_df['Duration_ms'].mean()

#KAROL G
kg_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'KAROL G']
avg_kg_duration = kg_df['Duration_ms'].mean()

#Mac DeMarco
mdm_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Mac DeMarco']
avg_mdm_duration = mdm_df['Duration_ms'].mean()

#The 1975
t1975_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'The 1975']
avg_t1975_duration = t1975_df['Duration_ms'].mean()

#Mitski
Mitski_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Mitski']
avg_Mitski_duration = Mitski_df['Duration_ms'].mean()

#Ozuna
Ozuna_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Ozuna']
avg_Ozuna_duration = Ozuna_df['Duration_ms'].mean()

#Hozier
Hozier_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Hozier']
avg_Hozier_duration = Hozier_df['Duration_ms'].mean()

#Lizzo
Lizzo_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Lizzo']
avg_Lizzo_duration = Lizzo_df['Duration_ms'].mean()

#League of Legends
lol_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'League of Legends']
avg_lol_duration = lol_df['Duration_ms'].mean()

#Bad Bunny
bb_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'Bad Bunny']
avg_bb_duration = bb_df['Duration_ms'].mean()

#BLACKPINK
BLACKPINK_df = spotify_youtube.loc[spotify_youtube['Artist'] == 'BLACKPINK']
avg_BLACKPINK_duration = BLACKPINK_df['Duration_ms'].mean()

song_duration_data = [['Metallica', avg_metallica_duration], ['Zendaya', avg_Zendaya_duration], ['Of Monsters and Men', avg_omm_duration], ['KAROL G', avg_kg_duration], ['Mac DeMarco', avg_mdm_duration],
                    ['The 1975', avg_t1975_duration], ['Mitski', avg_Mitski_duration], ['Ozuna', avg_Ozuna_duration], ['Hozier', avg_Hozier_duration], ['Lizzo', avg_Lizzo_duration], ['League of Legends',avg_lol_duration],
                    ['Bad Bunny',avg_bb_duration], ['BLACKPINK',avg_BLACKPINK_duration]]

song_duration_df = pd.DataFrame(song_duration_data, columns=['Artist', 'Average Song Duration']).sort_values('Average Song Duration',ascending=False)
```

```python
#scatter plot of duration times in ms
plt.bar( song_duration_df['Artist'], song_duration_df['Average Song Duration'])
plt.title("Average Song Duration per Artist in ms")
plt.xticks(rotation='vertical')
```
