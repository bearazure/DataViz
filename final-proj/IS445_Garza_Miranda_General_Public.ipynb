```python
# GENERAL PUBLIC VISUALIZATION 
# Dataset link -> https://www.kaggle.com/datasets/salvatorerastelli/spotify-and-youtube

import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Spotify-Youtube CSV into Dataframe! 
spotify_youtube = pd.read_csv("Spotify_Youtube.csv")
spotify_youtube
```

```python
#This part is a bit hardcoded.... if I could do something differently, I would make this more automated/less hardcoded

#1 Daft Punk 
daft_punk = spotify_youtube.loc[spotify_youtube['Artist'] == 'Daft Punk']
daft_punk

dp_avg_dance = daft_punk['Danceability'].mean()
print("Daft Punk Avg Danceability: ", dp_avg_dance)

#2 Pitbull 
pitbull = spotify_youtube.loc[spotify_youtube['Artist'] == 'Pitbull']
pitbull

pb_avg_dance = pitbull['Danceability'].mean()
print("Pitbull Avg Danceability: ", pb_avg_dance)

#3 Luis Miguel
luis_miguel = spotify_youtube.loc[spotify_youtube['Artist'] == 'Luis Miguel']
luis_miguel

lm_avg_dance = luis_miguel['Danceability'].mean()
print("Luis Miguel Avg Danceability: ", lm_avg_dance)

#4 MGMT
MGMT = spotify_youtube.loc[spotify_youtube['Artist'] == 'MGMT']
MGMT

MGMT_avg_dance = MGMT['Danceability'].mean()
print("MGMT Avg Danceability: ", MGMT_avg_dance)

#Zedd
zedd = spotify_youtube.loc[spotify_youtube['Artist'] == 'Zedd']
zedd

zedd_avg_dance = zedd['Danceability'].mean()
print("Zedd Avg Danceability: ", zedd_avg_dance)

#Wolfgang Amadeus Mozart
mozart = spotify_youtube.loc[spotify_youtube['Artist'] == 'Wolfgang Amadeus Mozart']
mozart

mozart_avg_dance = mozart['Danceability'].mean()
print("Wolfgang Amadeus Mozart Avg Danceability: ", mozart_avg_dance)

#Creating average danceability dataframe with select artists
data = [['Daft Punk', dp_avg_dance], ['Pitbull', pb_avg_dance], ['Luis Miguel', lm_avg_dance], 
        ['MGMT', MGMT_avg_dance], ['Zedd', zedd_avg_dance], ['Mozart', mozart_avg_dance]]
dance_df = pd.DataFrame(data, columns=['Artist', 'Average Danceability']).sort_values('Average Danceability', ascending=False)
```

```python
#bar chart -> My Favorite Artists and their Average Song Danceability
plt.bar(dance_df['Artist'], dance_df['Average Danceability'],color=['blue', 'red', 'green', 'purple', 'cyan', 'orange'])
plt.xlabel("Artist")
plt.ylabel("Average Danceability")
plt.title("My Favorite Artists and their Average Song Danceability")
```

```python
# Most popular songs by views 
top_300_songs = spotify_youtube.sort_values("Views", ascending=False).nlargest(300, 'Views')
top_100_songs = spotify_youtube.sort_values("Views", ascending=False).nlargest(100, 'Views')

#artist and how many times they appear on the top n songs 
top_300_instances_df = pd.DataFrame(top_300_songs['Artist'].value_counts().reset_index().values, columns=["Artist", "Total Instances on Top 300"])
top_100_instances_df = pd.DataFrame(top_100_songs['Artist'].value_counts().reset_index().values, columns=["Artist", "Total Instances on Top 100"])
combined_df = pd.DataFrame()

#generate a list of all the 300 artists
artist_name_list = []
artist_names = top_300_instances_df['Artist']
for i in range(len(top_300_instances_df['Artist'])):
    artist_name_list.append(top_300_instances_df['Artist'][i])

#Combine instances DFs and replace NaaN with 0s 
combined_df = pd.concat([top_300_instances_df,top_100_instances_df["Total Instances on Top 100"]], axis=1)
combined_df['Total Instances on Top 100'].fillna(0, inplace = True)
combined_df['Total Instances on Top 300'].fillna(0, inplace = True)

#only focus on top 20
top_20_combined = combined_df.head(30)
```

```python
#Number of songs and their instances in both the top 300 and 100 based on Artist 
#Resource -> https://seaborn.pydata.org/examples/part_whole_bars.html

f, ax = plt.subplots(figsize=(15, 10))

sns.set_color_codes("pastel")
sns.barplot(y="Total Instances on Top 300", x="Artist", data=top_20_combined, label="Total Instances on Top 300", color="b")

sns.set_color_codes("muted")
sns.barplot(y="Total Instances on Top 100", x="Artist", data=top_20_combined, label="Total Instances on Top 100", color="b")

ax.legend(loc="center right", frameon=True)
ax.set(xlim=(0, 24), ylabel="", xlabel="Artists and their Frequency in Top 300 Songs")
sns.despine(left=True, bottom=True)

#Resource -> https://drawingfromdata.com/seaborn/matplotlib/visualization/rotate-axis-labels-matplotlib-seaborn.html 
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
```
