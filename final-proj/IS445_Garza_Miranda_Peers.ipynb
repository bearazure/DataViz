```python
# PEER VISUALIZATON 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

spotify_youtube = pd.read_csv("Spotify_Youtube.csv")
spotify_youtube
```

```python
# Most popular songs by views 
top_300_songs_df = spotify_youtube.sort_values("Views", ascending=False).nlargest(300, 'Views')
top_300_songs_df
```

```python
#I discovered that Seaborn has a color palette that is meant for colorblind people! Thought this was interesting. 
sns.relplot(x="Danceability", y='Energy',data=top_300_songs_df, size='Views', hue='Key', sizes=(40,400), alpha=0.5,height=6, palette='colorblind')
plt.title("Top 300 Most-Viewed Songs Breakdown")
```

```python
#likes, views, comments
#Reference -> https://seaborn.pydata.org/examples/scatterplot_sizes.html 

cmap = sns.cubehelix_palette(start=.5, rot=-.75, as_cmap=True)
g = sns.relplot(
    data=top_300_songs_df,
    x="Likes", y="Comments",
    hue="Key", size="Views",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
