```python
import pandas as pd 
from matplotlib import pyplot as plt

storeData = pd.read_csv("store_data.csv", encoding='windows-1252')

subcatDf = storeData['Sub-Category'].head(60)
profitDf = storeData['Profit'].head(60)

fig, ax = plt.subplots(figsize =(16, 9))
ax.barh(subcatDf, profitDf)
ax.invert_yaxis()
ax.set_title('Top Sub-Categories Profits')
ax.grid(color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
```
